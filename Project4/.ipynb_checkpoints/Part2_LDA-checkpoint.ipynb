```python
tokenized_tweets = []
for tweet in clean_tweets:
    tokenized_tweet = nlp(tweet)
    
    tweet = "" # we want to keep each tweet seperate
    
    for token in tokenized_tweet:
        if token.is_space:
            continue
        elif token.is_punct:
            continue
        elif token.is_stop:
            continue
        elif token.is_digit:
            continue
        elif len(token) == 1:
            continue
        elif len(token) == 2:
            continue
        else:
            tweet += str(token.lemma_) + " " #creating lemmatized version of tweet
        
    tokenized_tweets.append(tweet)
tokenized_tweets = list(map(str.strip, tokenized_tweets)) # strip whitespace
tokenized_tweets = [x for x in tokenized_tweets if x != ""]
```

## LDA Analysis

Tokenizing tweets

```python
gensim_format_tweets = []
for tweet in filtered_tweets:
    list_form = tweet.split()
    gensim_format_tweets.append(list_form)
```

Formatting for Gensim

To dictionary for LDA in Gensim

```python
gensim_dictionary = Dictionary(gensim_format_tweets)
```

filtering out extreme words (don't appear frequently, or too frequently)

```python
gensim_dictionary.filter_extremes(no_below=10, no_above=0.4)
gensim_dictionary.compactify() # remove gaps after words that were removed
```

```python

def bag_of_words_generator(lst, dictionary):
    assert type(dictionary) == Dictionary, "Please enter a Gensim Dictionary"
    for i in lst: 
        yield dictionary.doc2bow(i)
```

```python
MmCorpus.serialize(corpus_path+"{}.mm".format("Tweets"), bag_of_words_generator(gensim_format_tweets, gensim_dictionary))
```

```python
corpus = MmCorpus(corpus_path+"{}.mm".format("Tweets"))
```

```python
#words in our corpus
corpus.num_terms
```

### LDA Modeling

```python
lda = LdaMulticore(corpus, num_topics=10, minimum_probability=0.03, id2word=gensim_dictionary, passes=10)
```

```python
lda.print_topics()
```

```python
from wordcloud import WordCloud, STOPWORDS

import matplotlib.pyplot as plt
for t in range(lda.num_topics):
    plt.figure()
    plt.imshow(WordCloud().fit_words(lda.show_topic(t, 200)))
    plt.axis("off")
    plt.title("Topic #" + str(t))
    plt.show()
```

### A Different Approach 

```python
# Create a CountVectorizer for parsing/counting words
count_vectorizer = CountVectorizer(ngram_range=(1, 2), max_df=.75,
                                   stop_words=nltk_stopwords, token_pattern="\\b[a-z][a-z]+\\b")
count_vectorizer.fit(tokenized_tweets)
```

```python
# Create the term-document matrix
# Transpose it so the terms are the rows
counts = count_vectorizer.transform(tokenized_tweets).transpose()
```

```python
id2word = dict((v, k) for k, v in count_vectorizer.vocabulary_.items())
```

```python
corpus = matutils.Sparse2Corpus(counts)
```

```python
lda = models.LdaModel(corpus=corpus, num_topics=10, id2word=id2word, passes=10)
```

```python
lda.print_topics()
```

```python
from wordcloud import WordCloud, STOPWORDS

import matplotlib.pyplot as plt
for t in range(lda.num_topics):
    plt.figure()
    plt.imshow(WordCloud().fit_words(lda.show_topic(t, 200)))
    plt.axis("off")
    plt.title("Topic #" + str(t))
    plt.show()
```
