```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
import seaborn as sns
from datetime import datetime

%matplotlib inline
```

```python
df = pd.read_csv('MTA_SPRING2017.csv') ##load data and create dataframe
```

```python
df2 = df[np.abs(df.ENTRIES-df.ENTRIES.mean())<=(3*df.ENTRIES.std())] #keep only the ones that are within +3 to -3 standard deviations in the column 'Data'.
```

```python
df2 = df2.sort_values('ENTRIES', ascending=False) ## check head for any outliers
df2
```

```python
df3 = df2.iloc[45:] ## remove outliers skewing data
```

```python
df3.head(10)
```

```python
df_sums = df3.groupby('STATION').sum() ## Create a new column for 'Station' 
df_sums['STATION'] = df_sums.index
df_sums.head(10)
```

```python
df_sums.sort_values('ENTRIES', ascending=False).head(15)
```

```python
df_sum_top = df_sums.sort_values('ENTRIES', ascending=False).head(15) ##Create a new dataframe storing most popular stations 
df_sum_top
```

```python
bar1 = sns.barplot(data = df_sum_top, x='ENTRIES',y='STATION',estimator=sum) ##using top stations, creates barplot
plt.setp(bar1.get_xticklabels(), rotation=0)
plt.xlabel('Entries (tens of millions)', fontsize=18)
plt.ylabel('Station', fontsize=18);
```

```python
df3["DateTime"] = pd.to_datetime(df3["DATETIME"]) ## converts panda series into datetime 
df3['just_date'] = df3['DateTime'].dt.date ## using our new datetime object, extracts the date only
```

```python
GRD_CNTRL = df3.loc[df['STATION'] == 'GRD CNTRL-42 ST'] ## new dataframe which filters all entries for Grand Central
```

```python
GRD_CNTRL = GRD_CNTRL.sort_values('ENTRIES', ascending=False)
```

```python
##GRD_CNTRL = GRD_CNTRL.iloc[6:] #edited since .ix is now deprecated.
GRD_WEEKLY = GRD_CNTRL.groupby(GRD_CNTRL.DateTime.dt.dayofweek).sum()
```

```python
GRD_WEEKLY.head(7)
```

```python
bar1 = sns.barplot(data = GRD_WEEKLY, x=GRD_WEEKLY.index,y='ENTRIES',estimator=sum)
plt.setp(bar1.get_xticklabels(), rotation=0)
plt.xlabel('Day of Week', fontsize=18)
```

```python
GRD_HOURLY = GRD_CNTRL.groupby(GRD_CNTRL.DateTime.dt.hour).sum() ##Group all rows according to the hour recorded
GRD_HOURLY.head(24)
```

```python
bar1 = sns.barplot(data = GRD_HOURLY, x=GRD_HOURLY.index,y='ENTRIES') ## plots entries to show most trafficked hours
plt.setp(bar1.get_xticklabels(), rotation=0)
plt.xlabel('Hour', fontsize=18);
```

```python
plt.plot((GRD_HOURLY['ENTRIES'])) ## now plot in line graphy to show general trends
plt.show()
```
