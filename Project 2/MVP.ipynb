```python
df = pd.read_csv('./data/Obese_Model_data.csv')
```

```python
df.sort_values('GDPper_cap', ascending=False)
```

```python
fig, ax = plt.subplots()
ax.set_xlabel('Education')
ax.set_ylabel('Percent Obese')
m, b = np.polyfit(df.Education, df.Percent_Obese, 1)
plt.plot(df.Education, df.Percent_Obese, '.')
plt.plot(df.Education, m*df.Education + b, '-');
```

```python
fig, ax = plt.subplots()
sns.set_style("darkgrid")
ax.set_xlabel('GDPper_cap')
ax.set_ylabel('Percent Obese')
m, b = np.polyfit(df.GDPper_cap, df.Percent_Obese, 1)
plt.plot(df.GDPper_cap, df.Percent_Obese, '.')
plt.plot(df.GDPper_cap, m*df.GDPper_cap + b, '-');
fig.savefig('GDP Per Cap.png', bbox_inches='tight')
```
