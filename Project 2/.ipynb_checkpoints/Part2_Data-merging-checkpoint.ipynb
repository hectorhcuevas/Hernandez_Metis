```python
#loading in new dataset relating to levels of insufficient activity
dfi = pd.read_csv('Countries_inactivity.csv')
```

```python
dfi.head()
```

```python
dfi.shape
```

```python
#Creating act variable to use mothods/ loops on and get rid of 'no data', strings and periods
act = dfi['Insufficiently active']
```

```python
clean_act = []
for x in act:
    clean_act.append(x[:4])
```

```python
[x.strip() for x in clean_act]
```

```python
##Further cleaning/stripping
just_nums = [w.replace(' [', '') for w in clean_act]
```

```python
[x.strip() for x in just_nums]
```

```python
dfi['Insufficiently_active'] = just_nums
```

```python
#filtering down to country and dependent variable 'insufficient activity'
dfi2 = dfi[['country','Insufficiently_active']]
```

```python
dfi2
```

```python
#Combinging our data by country column
df6 = pd.merge(df5,dfi2,on='country') ## merge the new dataframe to create new one of obesity&GDPs
```
