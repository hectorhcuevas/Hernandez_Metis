```python
# Create columns which we will fill with corresponding data from table 
Column_1 = soup.find_all(class_='column-1')
Column_2 = soup.find_all(class_='column-2')
Column_3 = soup.find_all(class_='column-3')
Column_4 = soup.find_all(class_='column-4')
Column_5 = soup.find_all(class_='column-5')
Column_6 = soup.find_all(class_='column-6')
```

```python
c1 = []
c2 = []
c3 = []
c4 = []
c5 = []
c6 = []
```

```python
#Appending our data to the columns created
for row in Column_1:
    df_c1.append(row)
for row in Column_2:
    df_c2.append(row)
for row in Column_3:
    df_c3.append(row)
for row in Column_4:
    df_c4.append(row)
for row in Column_5:
    df_c5.append(row)
for row in Column_6:
    df_c6.append(row)
```
